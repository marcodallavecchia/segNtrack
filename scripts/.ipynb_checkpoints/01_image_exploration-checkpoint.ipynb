## Goal
The goal of this repository is to segment single nuclei and cells of the provided images and track them in time in order to extract fluorescence intensity information following certain stimulation performed during the experiment.

```python
# Define packages to import
import tifffile as tf
import numpy as np
import napari
from napari.utils import nbscreenshot
```

```python
# Define paths
data_path = '../data/'
```

```python
cells_name =  'xy1_cells.tif'
nuc_name = 'xy1_nuclei.tif'
cells = tf.imread(data_path+cells_name)
nuclei = tf.imread(data_path+nuc_name)

# our data are 1608x1608 images in time, with 20 frames.
print(cells.shape)
print(nuclei.shape)
```

### Remarks
As it can be seen from the preview below, this sample is not the easiest thing to segment:
1. Cells are quite clumped in islands (relatively typical of HEK293T cells), accentuated by the staining treatment they  have undergone through
2. Cells move between frames (each frame was acquired at 3 minutes intervals)
3. At around frame 5 a blob of presumably dead cells float in the FOV
4. Fluorescent cells are only the ones that have been successfully transfect, but there are  many more that are not visible, meaning that there is no 1-to-1 correspondence between cells and nuclei (stained with Hoechst).

```python
# Play around with napari viewer to explore the data
v = napari.Viewer() # create viewer
# add both channels to viewer adjusting colormap and contrast limits for better visualization
v.add_image(cells, name='cell', colormap='green', contrast_limits=(cells.min(),cells.max()*0.7))
v.add_image(nuclei, name='nuclei', blending='additive', colormap='blue', contrast_limits=(nuclei.min(),nuclei.max()*0.7))
nbscreenshot(v) # add screenshot to notebook
```
